```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("nameContainer","unit-catalog")
dbutils.widgets.text("nameStorage","adlspchasiquiza01")
dbutils.widgets.text("catalogo","catalog_finalproject")
```

```sql
%python

nameContainer = dbutils.widgets.get("nameContainer")
nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta = f"abfss://{nameContainer}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.orders_bz;
DROP TABLE IF EXISTS ${catalogo}.bronze.customers_bz;
DROP TABLE IF EXISTS ${catalogo}.bronze.sales_bz;
DROP TABLE IF EXISTS ${catalogo}.bronze.products_bz;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta}/tablas/orders_bz", True)
dbutils.fs.rm(f"{ruta}/tablas/customers_bz", True)
dbutils.fs.rm(f"{ruta}/tablas/sales_bz", True)
dbutils.fs.rm(f"{ruta}/tablas/products_bz", True)
```
